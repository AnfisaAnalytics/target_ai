```python
import plotly.graph_objects as go
import plotly.io as pio
import numpy as np
import pandas as pd

# Установка SVG как рендерера по умолчанию
pio.renderers.default = "svg"

# 1. Простой линейный график
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines'))
fig.show()
```

```python
# 2. Scatter график с несколькими сериями
df = pd.DataFrame({
    'x': np.linspace(0, 10, 100),
    'sin': np.sin(np.linspace(0, 10, 100)),
    'cos': np.cos(np.linspace(0, 10, 100))
})

fig = go.Figure()
fig.add_trace(go.Scatter(x=df['x'], y=df['sin'], mode='lines', name='Синус'))
fig.add_trace(go.Scatter(x=df['x'], y=df['cos'], mode='lines', name='Косинус'))
fig.show()
```

```python

# 3. Гистограмма
data = np.random.randn(1000)
fig = go.Figure(data=[go.Histogram(x=data)])
fig.show()
```

```python

# 4. Диаграмма рассеивания с цветовой градацией
x = np.random.randn(500)
y = np.random.randn(500)
z = np.random.randn(500)

fig = go.Figure(data=go.Scatter(
    x=x, 
    y=y, 
    mode='markers',
    marker=dict(
        size=10,
        color=z,  # цвет маркеров зависит от значений z
        colorscale='Viridis',
        showscale=True
    )
))
fig.show()
```

```python
# 5. Сохранение графика в SVG файл
fig = go.Figure(data=go.Bar(x=['A', 'B', 'C'], y=[10, 20, 30]))
pio.write_image(fig, 'bar_chart.svg')
```
